Start by importing the data and splitting it by lines

```python
raw = []
with open("blaseball-log-season3.json", 'r') as f:
    raw += f.read().split("\n")
if(raw[-1] == ""):
    raw = raw[:-1]
```

```python
json_lines = []
for line in raw:
    json_lines.append(json.loads(line))
```
