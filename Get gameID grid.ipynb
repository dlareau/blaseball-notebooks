```python
import json
import glob
```

Start by importing the data and splitting it by lines

```python
filename = "season3.json"
#filename = "blaseball-log-season3.json"
#filename = "combined-blaseball-log-1596691883485.json"
```

```python
raw = []
with open(filename, 'r') as f:
    raw += f.read().split("\n")
if(raw[-1] == ""):
    raw = raw[:-1]
```

```python
json_lines = []
for line in raw:
    json_lines.append(json.loads(line))
```

How many lines did we get?

```python
print(len(json_lines))
```

```python
games = {}
for line in json_lines:
    game_ids = [update["_id"] for update in line["schedule"]]
    day = line["schedule"][0]["day"]
    if(day in games and games[day] != game_ids):
        print("mismatch")
    games[day] = game_ids
    
```
